## Importando os dados

```python
# selecionme as colunas loyalty_number, year, month, flights_booked, total_flights, distance e
# points_accumulated da tabela "flight_activity"
consulta_atividade = """
  SELECT
    loyalty_number,
    year,
    month,
    flights_booked,
    total_flights,
    distance,
    points_accumulated
  FROM
    flight_activity

"""
df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```

```python
# selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna distance for
# maior que 2000.
consulta_atividade = """
  SELECT
    loyalty_number,
    year,
    month,
    flights_booked,
    total_flights,
    distance,
    points_accumulated
  FROM
    flight_activity
  WHERE distance > 2000

"""
df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```

```python
# selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna distance for
# maior que 2000 e a coluna month for igual a 9.
consulta_atividade = """
  SELECT
    loyalty_number,
    year,
    month,
    flights_booked,
    total_flights,
    distance,
    points_accumulated
  FROM
    flight_activity
  WHERE distance > 2000 AND month == 9

"""
df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```

```python
# selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna distance for
# maior que 2000 ou a coluna points_accumulated for menor que 100.
consulta_atividade = """
  SELECT
    loyalty_number,
    year,
    month,
    flights_booked,
    total_flights,
    distance,
    points_accumulated
  FROM
    flight_activity
  WHERE distance > 2000 OR points_accumulated < 100

"""
df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```

```python
# selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna loyalty_card for
#igual a Star da tabela "flight_loyalty_history".
consulta_atividade = """
  SELECT
    *
  FROM
    flight_loyalty_history
  WHERE loyalty_card == 'Star'

"""
df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```

```python
# EXERCÌCIO 2
# selecione as mesmas colunas, porém, recupere somente as linhas cuja coluna gender
# for igual a Female da tabela flight_loyalty_history
consulta_atividade = """
  SELECT
    loyalty_number,
    country,
    city,
    gender,
    loyalty_card,
    salary
  FROM
    flight_loyalty_history
  WHERE
    gender = 'Female'

"""
df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```

```python
# EXERCÌCIO 3
# selecione as colunas loyalty_number, year, month, distance,
# flights_booked e total_flights da tabela flight_activity, e recupere
# somente as linhas cuja coluna flights_booked for maior que 10 e a
# coluna flights_booked for menor que 12, da tabela flight_activity
consulta_atividade = """
  SELECT
    loyalty_number,
    year,
    month,
    distance,
    flights_booked,
    total_flights
  FROM
    flight_activity
  WHERE
     flights_booked > 10 AND flights_booked < 12

"""
df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```
